```
import pandas as pd
import numpy as np
```

```
url='https://raw.githubusercontent.com/Nikhil14384/Feature_Engineering/main/train_ohe.csv'
```

```
df=pd.read_csv(url,usecols=['X1','X2','X3','X4','X5','X6'])
```

```
df.head()
```

```
#lets check how namy labels in each column
for col in df:
  print(col,' : ',len(df[col].unique()),'labels')
```

lets see how many cols will get for the 6 cols we have if we perform ohe.



```
#ohe on 6 cols
pd.get_dummies(df,drop_first=True).shape
```

if we observe we got 117 cols just for 6 cols if we do ohe, just imagine for real life complex datasets. So we will be using a unique technique within ohe to get better results with very less columns.

```
#top 10 most frequent categories for X2 column
df.X2.value_counts().sort_values(ascending=False).head(20)
```

```
#make a list with the most frequent categories(10)
t10=[x for x in df.X2.value_counts().sort_values(ascending=False).head(10).index]
t10
```

```
#we makethe entire dataframe into o's and 1's
for label in t10:
  df[label]=np.where(df['X2']==label,1,0)
```

```
df[['X2']+t10].head(20)
```

Like this we can do for each and every column and ends up having 60 cols(10 for each). it will give the better results as well.
